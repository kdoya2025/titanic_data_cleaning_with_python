```python
df1=df1.drop_duplicates()
df1
```

```python
df1.info()
```

```python
df1['Ticket']=df1['Ticket'].str.replace(r'[^a-zA-Z0-9]','',regex= True)
```

```python
df1[['first_name','last_name']]=df1['Name'].str.split(',',n=2,expand=True)
```

```python
df1['Full_name']=df1['last_name'] + ' '+(df1['first_name'])
```

```python
df1=df1.drop(columns=['Name','first_name','last_name'])
```

```python
cols=list(df1.columns)
cols.insert(3,cols.pop(cols.index('Full_name')))
df1=df1[cols]
```

```python
df1=df1.drop(columns='Cabin')
```

```python
df1['Age']=df1['Age'].fillna(0).astype(int)
df1
```

```python
df1=df1.drop(df1[df1['Age'] == 0].index)
```

```python
for x in df1.index:
    if df1.loc[x,'Embarked']== '':
     df1.drop(x,inplace=True)
```

```python
df1=df1.dropna(subset=['Embarked'])
```

```python
df1.to_csv(r'C:\Users\USER\OneDrive\Documents\pandas tutoriel\personal_data_cleaning.csv', index=False)
```
